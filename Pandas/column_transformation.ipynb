```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("ESD.xlsx")

# Create new column GetBonus and add values according to conditions
df.loc[(df["Bonus %"] == 0),"GetBonus"] = "No Bonus"
df.loc[(df["Bonus %"]>0),"GetBonus"] = "Bonus"

print(df.head(10))
```

```python
# Combine first name and Job title 
df["Name and Designation"] = df["Full Name"].str.capitalize()+" "+df["Job Title"]
print(df.head(10))
```

```python

def extract(value):
    return value[0:1]

def  test_extract(value):
    return value.split()[0]

df["Sex"] = df["Gender"].map(extract)
df["First Name"] = df["Full Name"].map(test_extract)
\
print(df.head(10))
```

```python
# Group by find the aggregate or numbers accordingly

# returns no of employees in each department on basis of ID
gp = df.groupby("Department").agg({"EEID":"count"})
print(gp)
# returns no of employees in each department on basis of ID and seperate them in gender
gp_gender = df.groupby(["Department","Gender"]).agg({"EEID":"count"})
print(gp_gender)

# Returns avrage salary according to countries
gp_avg = df.groupby(["Country","Gender"]).agg({"Annual Salary":"max","Age":"mean"})
print(gp_avg)
```
